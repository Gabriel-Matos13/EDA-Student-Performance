# 1) Importando librer√≠as necessarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuracion Visual
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)
```

```python
from google.colab import files
uploaded = files.upload()


```

```python
df = pd.read_csv(next(iter(uploaded)))
df.head()
```

# 3) Informaci√≥n general del dataset

```python
print("Informaci√≥n general del dataset:")
print(df.info())

print("\nEstad√≠sticas descriptivas:")
print(df.describe(include='all'))
```

#  4) Limpieza de datos

```python
print("\n¬øValores nulos por columna?")
print(df.isnull().sum())

print("\n¬øHay filas duplicadas?")
print(df.duplicated().sum())
```

# 5) An√°lisis univariado (distribuciones)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['GPA'], kde=True, bins=30, color='skyblue')
plt.title('Distribuci√≥n de GPA')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(df['StudyTimeWeekly'], kde=True, bins=30, color='green')
plt.title('Horas de estudio semanal')
plt.show()
```

```python
# An√°lisis de variables categ√≥ricas
plt.figure(figsize=(8, 4))
sns.countplot(x='Gender', data=df)
plt.title('Distribuci√≥n por g√©nero')
plt.show()

plt.figure(figsize=(12, 4))
sns.countplot(x='ParentalEducation', data=df, order=df['ParentalEducation'].value_counts().index)
plt.title('Nivel educativo de los padres')
plt.xticks(rotation=45)
plt.show()
```

#  6) An√°lisis bivariado


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='GPA', data=df)
plt.title('GPA por G√©nero')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', hue='Gender', data=df)
plt.title('Relaci√≥n entre Horas de Estudio y GPA')
plt.show()
```

#Correlaci√≥n entre variables num√©ricas

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n')
plt.show()
```

# **Conclusiones Generales del An√°lisis Exploratorio de Datos (EDA)**

**Dataset: Rendimiento acad√©mico de estudiantes (n = 2,392)**

El an√°lisis exploratorio realizado sobre el conjunto de datos ha revelado importantes patrones y relaciones entre variables que permiten comprender mejor los factores que influyen en el desempe√±o estudiantil. A continuaci√≥n se presentan las principales conclusiones:

* **1. Calidad de los datos**

El dataset est√° completamente limpio. No se identificaron valores nulos ni registros duplicados. Esto facilita la fiabilidad de los an√°lisis sin necesidad de tratamientos de datos previos.

*  **2. Desempe√±o acad√©mico**

El rendimiento promedio de los estudiantes, medido a trav√©s del GPA, es bajo (media ‚âà 1.90 sobre 4.00). La distribuci√≥n del GPA es asim√©trica, con la mayor√≠a de los estudiantes concentrados entre 1.5 y 2.5 puntos. Solo una minor√≠a alcanza promedios superiores a 3.5.

* **3. Ausentismo escolar como factor cr√≠tico**

La variable m√°s fuertemente correlacionada con el rendimiento acad√©mico es el n√∫mero de ausencias (Absences), que tiene:

*  Una correlaci√≥n de -0.92 con el GPA, lo que indica una relaci√≥n negativa extremadamente fuerte.

* Una correlaci√≥n de 0.73 con GradeClass, lo cual sugiere que un mayor n√∫mero de ausencias se asocia con una peor clasificaci√≥n acad√©mica (si se asume que 0 es mejor y 4 es peor).

Esto confirma que el ausentismo escolar es el principal predictor del bajo rendimiento acad√©mico en este conjunto de datos.

* **4. Relaci√≥n entre GPA y GradeClass**

Contrario a lo esperado, la correlaci√≥n entre el GPA y la categor√≠a acad√©mica (GradeClass) es de -0.78, lo cual indica que a medida que el GPA aumenta, el valor de GradeClass disminuye.

* Interpretaci√≥n: Esto sugiere que GradeClass est√° codificada inversamente (por ejemplo: 0 = Excelente, 4 = Deficiente), por lo tanto, un GPA alto corresponde a una mejor categor√≠a acad√©mica (n√∫mero m√°s bajo en GradeClass), lo que valida la consistencia del dataset.

* **5. Otros factores influyentes**
Aunque con menor fuerza que las ausencias, otras variables tambi√©n presentan relaciones notables:

* StudyTimeWeekly (horas de estudio semanal) tiene una correlaci√≥n positiva de 0.36 con el GPA: m√°s horas de estudio est√°n asociadas con un mejor desempe√±o.

* ParentalSupport tiene una correlaci√≥n positiva leve (0.19) con el GPA, lo cual resalta la importancia del entorno familiar.

* Tutoring muestra una relaci√≥n d√©bil pero positiva (0.15), lo que sugiere que contar con tutor√≠a podr√≠a apoyar el rendimiento, aunque no de forma determinante.

* **6. Participaci√≥n extracurricular**

Variables como Sports, Music, Extracurricular y Volunteering muestran correlaciones cercanas a cero con el GPA y GradeClass, indicando que su impacto en el rendimiento acad√©mico no es lineal o es muy bajo.



#  **Reflexi√≥n Final**

Los resultados muestran que el factor m√°s influyente en el rendimiento acad√©mico es el ausentismo, seguido por las horas de estudio y el apoyo familiar. Tambi√©n se identific√≥ que el GradeClass parece representar una escala inversa de desempe√±o, dado su comportamiento negativo frente al GPA.

Este an√°lisis permite establecer que para mejorar el rendimiento estudiantil, las estrategias deben centrarse en:

* Reducir el n√∫mero de ausencias.

* Fomentar h√°bitos efectivos de estudio.

* Impulsar el apoyo familiar y el acceso a tutor√≠as.
